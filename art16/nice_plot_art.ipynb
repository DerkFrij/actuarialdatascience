# Nice plot art
The art of creating a nice plot

*Note:* <br>To see the plot in Tableau style, first store the stylesheet "tableau.mplstyle" in your configdir

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Find out where to store the stylesheet
matplotlib.get_configdir()
```

```python
# read data
xlswb = 'data_zero_bond_yield_curves.xlsx'
data = pd.read_excel(xlswb, index_col='year')
```

```python
# create plot
plt.style.use('tableau')
fig, ax = plt.subplots()
fig.set_dpi(720)
ax.plot(data.index, data['Zero jan 2018'])
ax.plot(data.index, data['Zero dec 2018'])
ax.set_title('In 2018 pension funds suffered from decreasing rates')
ax.set_xlabel('maturity (years)')
ax.set_ylabel('rate (%)')
ax.set_xlim(xmin=0)
ax.set_ylim(ymin=-0.5, ymax=2.75)
ax.annotate(s='January', xy=(82, 2.35), color='tab:blue', size=8)
ax.annotate(s='December', xy=(82, 1.8), color='tab:orange', size=8)
```
